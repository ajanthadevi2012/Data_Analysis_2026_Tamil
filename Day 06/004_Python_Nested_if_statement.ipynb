# Python Nested `if` statement



### Syntax:

#### Example 1:
```python
if conditon_outer:
    if condition_inner:
        statement of nested if
    else:
        statement of nested if else:
    statement ot outer if
else:
    Outer else 
statement outside if block
```

#### Example 2:
```python
if expression1:
    statement(s)
    if expression2:
        statement(s)
    elif expression3:
        statement(s)
    elif expression4:
        statement(s)
    else:
        statement(s)
else:
    statement(s)
```

````python
x=10; y=12

if x>y:
    print("x is greater")
elif x<y:
    print("y is greater")
    if x==10:
        print("valid")
    else:
        print("hi")
else:
    print("x==y")
    
````
