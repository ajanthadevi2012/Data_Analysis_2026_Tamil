```python
df_2 = pd.read_csv('gun-violence-data1.csv')
```

```python
df_2.head()
```

```python
df_2.shape
```

```python
from datetime import datetime

#df_2['date'] = pd.to_datetime(df_2['date'])
df_2['date'] = pd.to_datetime(df_2['date'],dayfirst = True)

df_2['day'] = df_2['date'].dt.day
df_2['month'] = df_2['date'].dt.month
df_2['year'] = df_2['date'].dt.year
df_2['day_of_week'] =df_2['date'].dt.dayofweek

df_2['tot_victim'] = df_2['n_killed']+df_2['n_injured']
```

```python
df_2.head()
```

```python
df_2['day_of_week']  = df_2['day_of_week'].map({0:"monday",1:"tue",2:"wed",3:'thur',4:"fri",5:"sat",6:"Sunday"})
```

```python
df_2.head()
```

```python
df_2.info()
```

```python
df_2['day_of_week'].value_counts()
```

```python
# groupby
df_2.groupby('day_of_week')['tot_victim'].count()
```

```python
df_2.groupby('state')['tot_victim'].sum()
```

```python
df_2.groupby('state')['tot_victim'].sum().sort_values(ascending = True)
```

```python
df_2.groupby('month')['tot_victim'].count().sort_values(ascending = False)
```

```python
# combining 2 dataframe

df1 = pd.DataFrame({"name":["Ajantha","devi"],"age":[60,70]})
df2 = pd.DataFrame({"name":["steve","jobs"],"age":[56,67]})
```

```python
df1
```

```python
df2
```

```python
# concat
pd.concat([df1,df2],axis = 1)
```

```python
df3 = pd.concat([df1,df2],axis = 0)
df3
```

```python
df4 = pd.DataFrame({"name":["vicky","micky"],"age":[67,78],"hobby":["swimming","reading"]})
df4
```

```python
pd.concat([df3,df4],join = 'outer', ignore_index=True)
```

```python
pd.concat([df3,df4],join = 'inner', ignore_index=True)
```

```python
# merge
size = pd.DataFrame({"colour":['red','blue','yellow'],"size":['S','M',"L"]})
gender = pd.DataFrame({"colour":['red','green','yellow'],"gender":['F','M',"F"]})
```

```python
size
```

```python
gender
```

```python
pd.merge(size,gender,how = 'outer')
```

```python
pd.merge(size,gender,how = 'inner')
```

```python
pd.merge(size,gender,how = 'left')
```

```python
pd.merge(size,gender,how = 'right')
```
