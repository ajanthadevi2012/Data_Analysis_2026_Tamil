```python
# multi
class shape:
    def set_shape(self, name, area):
        self.shape_name = name
        self.shape_area = area
    def display_shape(self):
        print("the shape name is :", self.shape_name)
        print("the area of shape is : ",self.shape_area)
```

```python
class circle(shape):
    def set_circle(self, radius):
        self.radius = radius
    def display_circle(self):
        print("the radius of circle is :", self.radius)
```

```python
cir1 = circle()
```

```python
cir1.set_shape('circle',200)
cir1.display_shape()
```

```python
cir1.set_circle(50)
cir1.display_circle()
```

```python
class sq(shape):
    def set_sq(self, sides):
        self.sides = sides
    def display_sq(self):
        print("the radius of circle is :", self.sides)
```

```python
sq1 = sq()
```

```python
# multi = > level
class sq_1(circle):
    def set_sq_1(self, sides):
        self.sides = sides
    def display_sq_1(self):
        print("the radius of circle is :", self.sides)
```
